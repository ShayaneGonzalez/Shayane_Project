##Matriz de sensibilidade cheia para o caso em que o dado observado é gz

```python
def AZ(xi,yi,zi,xk,yk,zk):
    raio = 1.0
    volume = 1.0 
    rz = zi-zk
    rx = xi-xk
    ry = yi-yk
    r  = np.sqrt(rx**2+ry**2+rz**2)
    r3=r**3
    gz = -volume*G*rz/r3
    Az = gz*SI2MGAL
    return Az
```

##Matriz diagonal Ã

```python
diagonal_A = gama*2.*np.pi/ds2
Anew = diags(diagonal_A, 0, (N,N), format='csr', dtype='float')
for i in range(len(xi)):
    if(5100.<=yi[i]<=7300.):
        Anew[i,i]=gama*2.*np.pi/ds1
```

##Chute inicial para a densidade e dado predito com essa densidade

```python
rho0 = linalg.spsolve(Anew,do)
```

```python
rho = np.copy(rho0)
A_cheia = np.empty((N,M),dtype =float)
for j  in range(M):
    A_cheia[:,j] =AZ(xi,yi,zi,xi[j],yi[j],zk[j])
gzpre = np.dot(A_cheia,rho)
```

#Estimando p* de uma maneira iterativa

```python
start_time_est_p = datetime.now()
```

```python
ITMAX = 30
phi = []
relative_variation = []
it = []
alpha = 0.00001
beta = 0.001
for i in range(ITMAX):
    gzpre = np.dot(A_cheia,rho)
    res = do - gzpre
    drho = linalg.spsolve(Anew,res)
    a = np.sum(rho*rho)
    b = np.sum(drho*drho)
    rho += drho
    phi.append(np.sum(res*res)/N)
    relative_variation.append(b/a)
    it.append(i)
    print 'i',i,'phi = norma res /N',phi[i]
    print 'i',i,'Relative Variation norma rho/ norma delta rho',relative_variation[i]
LU_flops = (2./3.)*N**3 + 2.*N**2
ItEqlayer_flops = N + (it[-1]+1.)*(10.*N + 2.*N**2)

flops_ratio = LU_flops/ItEqlayer_flops

print 'LU_flops:        %10.5e' % LU_flops
print 'ItEqlayer_flops: %10.5e' % ItEqlayer_flops
print 'flops_ratio:     %.1f' % flops_ratio

rmean = res.mean()
rstd = res.std()
print "Residuals"
print "mean:   %10.5f mGal" % rmean
print "stddev: %10.5f mGal" % rstd
```

```python
end_time_est_p = datetime.now()
print('Duration: {}'.format(end_time_est_p - start_time_est_p))
```

##Início do filtro móvel

```python
tamanho= 80.
rho2=np.zeros(len(rho))
for i in range(len(rho)):
    soma=0.
    n=0
    for j in range(len(xi)):
        if xi[j]<(xi[i]+tamanho) and xi[j]>(xi[i]-tamanho) and yi[j]<(yi[i]+tamanho) and yi[j]>(yi[i]-tamanho):
            soma=soma+rho[j]
            n=n+1
    rho2[i]=soma/n
```

#Cálculo da componente horizontal  gx

```python
def AX(xi,yi,zi,xk,yk,zk):
    raio = 1.0
    volume = 1.0 
    rz = zi-zk
    rx = xi-xk
    ry = yi-yk
    r  = np.sqrt(rx**2+ry**2+rz**2)
    r3=r**3
    gx = -volume*G*rx/r3
    Ax = gx*SI2MGAL
    return Ax
```

```python
gx_calc=[]

for j  in range(M):
    gx_calc=np.append(gx_calc, np.dot(AX(xi[j],yi[j],zi[j],xi,yi,zk), rho2))

gx_real=ftg[1]
res_gx=gx_real-gx_calc
```

#Cálculo da componente horizontal  gy

```python
def AY(xi,yi,zi,xk,yk,zk):
    raio = 1.0
    volume = 1.0 
    rz = zi-zk
    rx = xi-xk
    ry = yi-yk
    r  = np.sqrt(rx**2+ry**2+rz**2)
    r3=r**3
    gy = -volume*G*ry/r3
    Ay = gy*SI2MGAL
    return Ay
```

```python
gy_calc=[]

for j  in range(M):
    gy_calc=np.append(gy_calc, np.dot(AY(xi[j],yi[j],zi[j],xi,yi,zk), rho2))

gy_real=ftg[2]
res_gy=gy_real-gy_calc
```
