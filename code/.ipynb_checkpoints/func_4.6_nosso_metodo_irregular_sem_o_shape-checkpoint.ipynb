##Matriz A

```python
def A(xi,yi,zi,xk,yk,zk):
    raio = 1.0
    volume = 1.0 
    rz = zi-zk
    rx = xi-xk
    ry = yi-yk
    r  = np.sqrt(rx**2+ry**2+rz**2)
    r3=r**3
    gx = -volume*G*rx/r3
    gy = -volume*G*ry/r3
    gz = -volume*G*rz/r3
    Ax = gx*SI2MGAL
    Ay = gy*SI2MGAL
    Az = gz*SI2MGAL
    return Ax, Ay, Az
```

```python
def AX(xi,yi,zi,xk,yk,zk):
    raio = 1.0
    volume = 1.0 
    rz = zi-zk
    rx = xi-xk
    ry = yi-yk
    r  = np.sqrt(rx**2+ry**2+rz**2)
    r3=r**3
    gx = -volume*G*rx/r3
    gy = -volume*G*ry/r3
    gz = -volume*G*rz/r3
    Ax = gx*SI2MGAL
    return Ax
```

```python
def AY(xi,yi,zi,xk,yk,zk):
    raio = 1.0
    volume = 1.0 
    rz = zi-zk
    rx = xi-xk
    ry = yi-yk
    r  = np.sqrt(rx**2+ry**2+rz**2)
    r3=r**3
    gx = -volume*G*rx/r3
    gy = -volume*G*ry/r3
    gz = -volume*G*rz/r3
    Ay = gy*SI2MGAL
    return Ay
```

```python
def AZ(xi,yi,zi,xk,yk,zk):
    raio = 1.0
    volume = 1.0 
    rz = zi-zk
    rx = xi-xk
    ry = yi-yk
    r  = np.sqrt(rx**2+ry**2+rz**2)
    r3=r**3
    gx = -volume*G*rx/r3
    gy = -volume*G*ry/r3
    gz = -volume*G*rz/r3
    Az = gz*SI2MGAL
    return Az
```

###Iteração

```python
start_time_est_p = datetime.now()
```

```python
ITMAX = 30
rho = np.copy(rho0)
phi = []
relative_variation = []
it = []
alpha = 0.00001
beta = 0.001
for i in range(ITMAX):
    gzpre=[]
    for j in range(N):    
        gzpre = np.append(gzpre,np.dot(AZ(xi[j],yi[j],zi[j],xi,yi,zk),rho))
    w = do - gzpre
    drho = linalg.spsolve(Anew,w)
    a = np.sum(rho*rho)
    b = np.sum(drho*drho)
    rho += drho
    res = do-gzpre
    phi.append(np.sum(res*res)/N)
    relative_variation.append(b/a)
    it.append(i)
    print 'i',i,'phi = norma res /N',phi[i]
    print 'i',i,'Relative Variation norma rho/ norma delta rho',relative_variation[i]
    if (relative_variation[i] < alpha) or (phi[i] < beta):
        break
LU_flops = (2./3.)*N**3 + 2.*N**2
ItEqlayer_flops = N + (it[-1]+1.)*(10.*N + 2.*N**2)

flops_ratio = LU_flops/ItEqlayer_flops

print 'LU_flops:        %10.5e' % LU_flops
print 'ItEqlayer_flops: %10.5e' % ItEqlayer_flops
print 'flops_ratio:     %.1f' % flops_ratio

rmean = res.mean()
rstd = res.std()
print "Residuals"
print "mean:   %10.5f mGal" % rmean
print "stddev: %10.5f mGal" % rstd
```

```python
end_time_est_p = datetime.now()
print('Duration: {}'.format(end_time_est_p - start_time_est_p))
```

```python
def A_DERIVES(xi,yi,zi,xk,yk,zk):
    raio = 1.0
    volume = 1.0 
    size = len(xi)
    #for i in range(size):
    rz = zi-zk
    rx = xi-xk
    ry = yi-yk
    r  = np.sqrt(rx**2+ry**2+rz**2)
    r3=r**3
    r5=r**5
    gxz = volume*G*( (-2.*rx*rz)/r3  + 1.5*(rx/r5)  )
    gyz = volume*G*( (-2.*ry*rz)/r3  + 1.5*(ry/r5)  )
    gzz = volume*G*( (-1./r3)  + 3.*(rz*rz/r5)  )
    #Axz = gxz*SI2MGAL
    AXZ = gxz*SI2EOTVOS
    AYZ = gxz*SI2EOTVOS
    AZZ = gxz*SI2EOTVOS
    return AXZ, AYZ, AZZ
```

```python
delta_z=1.0
for j  in range(M):
    AX[:,j], AY[:,j], AZ[:,j] =A(xi,yi,zi+delta_z,xi[j],yi[j],zk[j])  
    
gzpre_dz = np.dot(AZ,rho)

#gzpre_dz = sphere.gz(xi,yi,zi+delta_z, layer)
```
