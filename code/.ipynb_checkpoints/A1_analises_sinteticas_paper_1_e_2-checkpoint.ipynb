#Estimando p* de uma maneira iterativa

###Constantes

```python
gama=G*SI2MGAL
```

##Matriz A

```python
def AX(xi,yi,zi,xk,yk,zk):
    raio = 1.0
    volume = 1.0 
    rz = zi-zk
    rx = xi-xk
    ry = yi-yk
    r  = np.sqrt(rx**2+ry**2+rz**2)
    r3=r**3
    gx = -volume*G*rx/r3
    Ax = gx*SI2MGAL
    return Ax
```

```python
def AY(xi,yi,zi,xk,yk,zk):
    raio = 1.0
    volume = 1.0 
    rz = zi-zk
    rx = xi-xk
    ry = yi-yk
    r  = np.sqrt(rx**2+ry**2+rz**2)
    r3=r**3
    gy = -volume*G*ry/r3
    Ay = gy*SI2MGAL
    return Ay
```

```python
def AZ(xi,yi,zi,xk,yk,zk):
    raio = 1.0
    volume = 1.0 
    rz = zi-zk
    rx = xi-xk
    ry = yi-yk
    r  = np.sqrt(rx**2+ry**2+rz**2)
    r3=r**3
    gz = -volume*G*rz/r3
    Az = gz*SI2MGAL
    return Az
```

##Matriz Ã

```python
dy1, dx1 = gridder.spacing(area1, shape1)
ds1=dx1*dy1
print 'dx1 = %.5f' % dx1
print 'dy1 = %.5f' % dy1
print 'ds1 = %.5f' % ds1
```

```python
dy2, dx2 = gridder.spacing(area2, shape2)
ds2=dx2*dy2
print 'dx2 = %.5f' % dx2
print 'dy2 = %.5f' % dy2
print 'ds2 = %.5f' % ds2
```

```python
diagonal_A = gama*2.*np.pi/ds2
Anew = diags(diagonal_A, 0, (N,N), format='csr', dtype='float')
for i in range(len(xi)):
    if(5100.<=yi[i]<=7300.):
        Anew[i,i]=gama*2.*np.pi/ds1
```

##Chute inicial para a densidade e dado predito com essa densidade

```python
rho0 = linalg.spsolve(Anew,do)
```

```python
rho = np.copy(rho0)

A_cheia = np.empty((N,M),dtype =float)

for j  in range(M):
    A_cheia[:,j] =AZ(xi,yi,zi,xi[j],yi[j],zk[j])
    
gzpre = np.dot(A_cheia,rho)
```

###Iteração

```python
start_time_est_p = datetime.now()
```

```python
ITMAX = 30
phi = []
relative_variation = []
it = []
alpha = 0.00001
beta = 0.001
for i in range(ITMAX):
    gzpre = np.dot(A_cheia,rho)
    res = do - gzpre
    drho = linalg.spsolve(Anew,res)
    a = np.sum(rho*rho)
    b = np.sum(drho*drho)
    rho += drho
    phi.append(np.sum(res*res)/N)
    relative_variation.append(b/a)
    it.append(i)
    print 'i',i,'phi = norma res /N',phi[i]
    print 'i',i,'Relative Variation norma rho/ norma delta rho',relative_variation[i]
    #if (relative_variation[i] < alpha) or (phi[i] < beta):
    #    break
LU_flops = (2./3.)*N**3 + 2.*N**2
ItEqlayer_flops = N + (it[-1]+1.)*(10.*N + 2.*N**2)

flops_ratio = LU_flops/ItEqlayer_flops

print 'LU_flops:        %10.5e' % LU_flops
print 'ItEqlayer_flops: %10.5e' % ItEqlayer_flops
print 'flops_ratio:     %.1f' % flops_ratio

rmean = res.mean()
rstd = res.std()
print "Residuals"
print "mean:   %10.5f mGal" % rmean
print "stddev: %10.5f mGal" % rstd
```

```python
end_time_est_p = datetime.now()
print('Duration: {}'.format(end_time_est_p - start_time_est_p))
```

#Derivada calculada analiticamente

```python
def A_DERIVES(xi,yi,zi,xk,yk,zk):
    volume = 1.0 
    rx = xi-xk
    ry = yi-yk
    rz = zi-zk
    r  = np.sqrt(rx**2+ry**2+rz**2)
    r3=r**3
    r5=r**5
    
    gxx = volume*G*( (-1./r3)  + 3.*(rx*rx/r5) )
    gxy = volume*G*( (3.*rx*ry)/r5 )
    gxz = volume*G*( (3.*rx*rz)/r5 )
    gyy = volume*G*( (-1./r3)  + 3.*(ry*ry/r5) )
    gyz = volume*G*( (3.*ry*rz)/r5 )
    gzz = volume*G*( (-1./r3)  + 3.*(rz*rz/r5) )
    
    AXX = gxx*SI2EOTVOS
    AXY = gxy*SI2EOTVOS
    AXZ = gxz*SI2EOTVOS
    AYY = gyy*SI2EOTVOS
    AYZ = gyz*SI2EOTVOS
    AZZ = gzz*SI2EOTVOS
    
    return AXX, AXY, AXZ, AYY, AYZ, AZZ
```

```python
gxx_analitico = []
gxy_analitico = []
gxz_analitico = []
gyy_analitico = []
gyz_analitico = []
gzz_analitico = []

for j in range(M):
    gxx_analitico, gxy_analitico, gxz_analitico, gyy_analitico, gyz_analitico, gzz_analitico=np.append(gxx_analitico, np.dot(A_DERIVES(xi[j],yi[j],zi[j],xi,yi,zk)[0], rho)), np.append(gxy_analitico, np.dot(A_DERIVES(xi[j],yi[j],zi[j],xi,yi,zk)[1], rho)), np.append(gxz_analitico, np.dot(A_DERIVES(xi[j],yi[j],zi[j],xi,yi,zk)[2], rho)), np.append(gyy_analitico, np.dot(A_DERIVES(xi[j],yi[j],zi[j],xi,yi,zk)[3], rho)), np.append(gyz_analitico, np.dot(A_DERIVES(xi[j],yi[j],zi[j],xi,yi,zk)[4], rho)), np.append(gzz_analitico, np.dot(A_DERIVES(xi[j],yi[j],zi[j],xi,yi,zk)[5], rho))
```

```python
plt.figure(figsize=(15, 15))
mpl.subplot(3,3,1)
mpl.title("Gxx (Eotvos)")
plt.xlabel('Distancia (metros)')
plt.ylabel('Distancia (metros)')
mpl.axis('scaled')
levels = mpl.contourf(yi, xi, gxx_analitico, shape_obs, 12, interp=True)
mpl.colorbar()
mpl.subplot(3, 3, 2)
mpl.title("Gxy (Eotvos)")
plt.xlabel('Distancia (metros)')
plt.ylabel('Distancia (metros)')
mpl.axis('scaled')
levels = mpl.contourf(yi, xi, gxy_analitico, shape_obs, 12, interp=True)
mpl.colorbar()
mpl.subplot(3, 3, 3)
mpl.title("Gxz (Eotvos)")
plt.xlabel('Distancia (metros)')
plt.ylabel('Distancia (metros)')
mpl.axis('scaled')
#plt.plot(xi, yi, 'ko', ms=3)
levels = mpl.contourf(yi, xi, gxz_analitico, shape_obs, 12, interp=True)
mpl.colorbar()
mpl.subplot(3,3,5)
mpl.title("Gyy (Eotvos)")
plt.xlabel('Distancia (metros)')
plt.ylabel('Distancia (metros)')
mpl.axis('scaled')
levels = mpl.contourf(yi, xi, gyy_analitico, shape_obs, 12, interp=True)
mpl.colorbar()
mpl.subplot(3, 3, 6)
mpl.title("Gyz (Eotvos)")
plt.xlabel('Distancia (metros)')
plt.ylabel('Distancia (metros)')
mpl.axis('scaled')
levels = mpl.contourf(yi, xi, gyz_analitico, shape_obs, 12, interp=True)
mpl.colorbar()
mpl.subplot(3, 3, 9)
mpl.title("Gzz (Eotvos)")
plt.xlabel('Distancia (metros)')
plt.ylabel('Distancia (metros)')
mpl.axis('scaled')
#plt.plot(xi, yi, 'ko', ms=3)
levels = mpl.contourf(yi, xi, gzz_analitico, shape_obs, 12, interp=True)
mpl.colorbar()
mpl.savefig('ftg_sintetica_analitica_paper_2.png', dpi=600, pad_inches=0.0)
mpl.show()
```
