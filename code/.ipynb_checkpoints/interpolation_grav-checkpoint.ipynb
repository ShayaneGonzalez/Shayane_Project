# Interpolation of gravity data

### Brief description

This Jupyter notebook is a supplementary material contaning the code needed to generate part of the synthetic data and reproduce part of the figures and results of the manuscript. This code uses the open-source Python toolkit for geophysical modeling and inversion [Fatiando a Terra](http://www.fatiando.org/index.html).

This code presents the application of the equivalent layer technique for interpolating gravity anomaly data.

### Observed gravity anomaly

Gravity anomaly produced by the synthetic prisms (Figure 4) at the observation points (black dots) located on the undulating surface shown in Figure 13. This anomaly is calculated with equation 1 and is contaminated by a pseudo-random Gaussian noise having null mean and standard deviation equal to 1 mGal.

```python
gz = contaminate(prism.gz(xp,yp,zp,full_model['model']), 1.0, seed=32)
```

### Planar equivalent layer

We use a planar equivalent layer formed by one point mass placed beneath each observation point at a depth of $z_{0} = -200$ m. In this case, the harmonic functions $g_{ij}$ are defined according to equation 63.

#### Parameters defining the equivalent layer

```python
#horizontal plane containing the equivalent sources
z0 = np.zeros_like(zp)
#h = np.max(zp) + 1.4*dx
h = -200.
z0 += h
```

```python
print h
```

```python
#equivalent layer
layer = PointGrid(airborne_grid['area'], h, airborne_grid['shape'])
```

#### Linear inverse problem for estimating the physical property distribution within the equivalent layer

```python
# misfit function
misfit = EQLGravity(xp, yp, zp, gz, layer)
```

```python
# normalizing factor
f0 = np.trace(misfit.hessian(None))/misfit.nparams
```

```python
print 'f0 = %10.3e' % f0
```

```python
# Tikhonov regularization
#regul = Damping(layer.size)
regul = Smoothness2D(layer.shape)
```

```python
# Use the L-curve to find the best regularization parameter
solver = LCurve(misfit, regul, [f0*(10.**i) for i in range(-10, -1)]).fit()
```

```python
print 'u0 = %10.3e' % ((1./f0)*solver.regul_param_)
```

```python
layer.addprop('density', solver.estimate_)
```

#### Data fit

```python
af.plotmap(xp, yp, gz, airborne_grid['shape'], airborne_grid['area'],
           color_scheme = 'Greys', label_title = 'mGal', 
           prism_projection = True, projection_style = '-k', 
           model = full_model['projection'], 
           figure_label = '(a)',
           figure_name = 'gz-noisy.png')

af.plotmap(xp, yp, solver[0].predicted(), airborne_grid['shape'], airborne_grid['area'],
           color_scheme = 'Greys', label_title = 'mGal', 
           prism_projection = True, projection_style = '-k', 
           model = full_model['projection'], 
           figure_label = '(b)',
           figure_name = 'gz-predicted.png')

af.plothist(solver[0].residuals(), file_name = 'hist-gz.png',
            text_fontsize = 12, figure_label = '(c)')
```

### Undulating equivalent layer

We use an undulating equivalent layer formed by one point mass placed beneath each observation point. In this case, the harmonic functions $g_{ij}$ are defined according to equation 63.

#### Parameters defining the equivalent layer

```python
print 'min  %.5f' % np.min(airborne_grid['spacing'])
print 'mean %.5f' % np.mean(airborne_grid['spacing'])
print 'max  %.5f' % np.max(airborne_grid['spacing'])
```

```python
#creates an undulating surface by using 2D Gaussian functions
#f, fx, fy = af.eqlayer_surface(xp, yp)
f = zp + 520.
```

```python
#verify if the undulating surface f is located below the
#observation surface (Figure 5)
np.alltrue(f > zp)
```

```python
af.plotmap(xp, yp, f, airborne_grid['shape'], airborne_grid['area'], 
           color_scheme = 'Greys_r', label_title = 'm', 
           prism_projection = True, projection_style = '-w', 
           model = full_model['projection'])
```

**Figure 8:** Undulating surface (grayscale) defining an equivalent layer. The black rectangles represent the horizontal projection of the simulated prisms (Figure 4).

```python
#equivalent layer
layer2 = PointGrid(airborne_grid['area'], f, airborne_grid['shape'])
```

#### Linear inverse problem for estimating the physical property distribution within the equivalent layer

```python
# misfit function
misfit2 = EQLGravity(xp, yp, zp, gz, layer2)
```

```python
# normalizing factor
f02 = np.trace(misfit2.hessian(None))/misfit2.nparams
```

```python
print 'f0 = %10.3e' % f02
```

```python
# Tikhonov regularization
#regul2 = Damping(layer2.size)
regul2 = Smoothness2D(layer2.shape)
```

```python
# Use the L-curve to find the best regularization parameter
solver2 = LCurve(misfit2, regul2, [f02*(10.**i) for i in range(-10, -1)]).fit()
```

```python
print 'u0 = %10.3e' % ((1./f02)*solver2.regul_param_)
```

```python
layer2.addprop('density', solver2.estimate_)
```

#### Data fit

```python
af.plotmap(xp, yp, solver2[0].predicted(), airborne_grid['shape'], airborne_grid['area'],
           color_scheme = 'Greys', label_title = 'mGal', 
           prism_projection = True, projection_style = '-k', 
           model = full_model['projection'], 
           figure_label = '(a)',
           figure_name = 'gz-predicted-undul.png')

#af.plothist(solver2[0].residuals(), file_name = 'hist-gz-undul.png', 
#            text_position = [0.6, 1.0, 0.93], text_fontsize = 12, figure_label = '(b)')
af.plothist(solver2[0].residuals(), file_name = 'hist-gz-undul.png', 
            text_fontsize = 12, figure_label = '(b)')
```
