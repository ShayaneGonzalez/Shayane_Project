###Constantes

```python
gama=G*SI2MGAL
```

##Matriz A

```python
def AX(xi,yi,zi,xk,yk,zk):
    raio = 1.0
    volume = 1.0 
    rz = zi-zk
    rx = xi-xk
    ry = yi-yk
    r  = np.sqrt(rx**2+ry**2+rz**2)
    r3=r**3
    gx = -volume*G*rx/r3
    Ax = gx*SI2MGAL
    return Ax
```

```python
def AY(xi,yi,zi,xk,yk,zk):
    raio = 1.0
    volume = 1.0 
    rz = zi-zk
    rx = xi-xk
    ry = yi-yk
    r  = np.sqrt(rx**2+ry**2+rz**2)
    r3=r**3
    gy = -volume*G*ry/r3
    Ay = gy*SI2MGAL
    return Ay
```

```python
def AZ(xi,yi,zi,xk,yk,zk):
    raio = 1.0
    volume = 1.0 
    rx = xi-xk
    ry = yi-yk
    rz = zi-zk
    r  = np.sqrt(rx**2+ry**2+rz**2)
    r3=r**3
    gz = -volume*G*rz/r3
    Az = gz*SI2MGAL
    return Az
```

```python
diagonal_A = gama*2.*np.pi/ds1
Anew = diags(diagonal_A, 0, (N,N), format='csr', dtype='float')
for i in range(len(xi)):
    if(-2500.<=yi[i]<=2500.):
        Anew[i,i]=gama*2.*np.pi/ds2
    if(2500.<=yi[i]<=5000.):
        Anew[i,i]=gama*2.*np.pi/ds3
```

###Iteração

```python
start_time_est_p = datetime.now()
```

```python
ITMAX = 30
phi = []
relative_variation = []
it = []
alpha = 0.00001
beta = 0.001
for i in range(ITMAX):
    gzpre = np.dot(A_cheia,rho)
    res = do - gzpre
    drho = linalg.spsolve(Anew,res)
    a = np.sum(rho*rho)
    b = np.sum(drho*drho)
    rho += drho
    phi.append(np.sum(res*res)/N)
    relative_variation.append(b/a)
    it.append(i)
    print 'i',i,'phi = norma res /N',phi[i]
    print 'i',i,'Relative Variation norma rho/ norma delta rho',relative_variation[i]
    #if (relative_variation[i] < alpha) or (phi[i] < beta):
    #    break
LU_flops = (2./3.)*N**3 + 2.*N**2
ItEqlayer_flops = N + (it[-1]+1.)*(10.*N + 2.*N**2)

flops_ratio = LU_flops/ItEqlayer_flops

print 'LU_flops:        %10.5e' % LU_flops
print 'ItEqlayer_flops: %10.5e' % ItEqlayer_flops
print 'flops_ratio:     %.1f' % flops_ratio

rmean = res.mean()
rstd = res.std()
print "Residuals"
print "mean:   %10.5f mGal" % rmean
print "stddev: %10.5f mGal" % rstd
```

```python
def A_DERIVES(xi,yi,zi,xk,yk,zk):
    volume = 1.0 
    rx = xi-xk
    ry = yi-yk
    rz = zi-zk
    r  = np.sqrt(rx**2+ry**2+rz**2)
    r3=r**3
    r5=r**5
    gxz = volume*G*( (3.*rx*rz)/r5 )#volume*G*( (-2.*rx*rz)/r3  + 1.5*(rz/r5) )
    gyz = volume*G*( (3.*ry*rz)/r5 )#volume*G*( (-2.*ry*rz)/r3  + 1.5*(rz/r5) )
    gzz = volume*G*( (-1./r3)  + 3.*(rz*rz/r5) )
    AXZ = gxz*SI2EOTVOS
    AYZ = gyz*SI2EOTVOS
    AZZ = gzz*SI2EOTVOS
    return AXZ, AYZ, AZZ
```
